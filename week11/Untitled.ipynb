```python
import pandas as pd
import numpy as np
import datetime as dt
data=pd.read_csv('meetings.csv')
data
```

Create a new column called startDate that contains just the day part of the startTime column in the data. You can do that using datetime functions if you want, but you'll notice that the date in this file is a string formatted YYYY-MM-DD-HH:mm:ss, so it's probably easier just to pull out the first 10 characters of the string instead. meetings['startTime'].str[0:10]

```python
#convert startTime column to a string
string_startDate=data['startTime'].astype(str)
string_startDate
```

```python
startDate=pd.to_datetime(string_startDate.str[0:10])
startDate
```

```python
data['startDate']=startDate
data.head()
```
