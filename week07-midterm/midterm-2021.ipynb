## Part 1: CHA2DS2-VASc

This scoring mechanism for Atrial Fibrillation Stroke uses 7 inputs:
* Age (Number)
* Sex (Male / Female)
* CHF History (True / False)
* Hypertension History (True / False)
* Stroke History (True / False)
* Vascular Disease History (True / False)
* Diabetes History (True / False)

Fill out the function below with logic to calculate the numeric risk score for teh given input.

Be sure to provide meaningful documentation and at least two test cases in your documentation.  Also make sure your code satisfies the test cases provided in the assert statements.

````python
def cha2ds2_vasc(age, sex, chf, hypertension, stroke, vascular, diabetes):
    """(int,str,bool,bool,bool,bool,bool)->int
    This function uses the logic from https://www.mdcalc.com/cha2ds2-vasc-score-atrial-fibrillation-stroke-risk
    to calculate the score Atrial Fibrillation Stroke.
    
    >>>>cha2ds2_vasc(22,'Male',False,False,True,False,False)
    2
    
    >>>>cha2ds2_vasc(32,'Female',True,True,True,True,True)
    7
    
    
    
    """
    ### YOUR SOLUTION HERE
   
    
    
    score=0
    
    
    if age < 65:
        score += 0
    elif (age >=65) and (age <=74):
        score += 1
    elif age >= 75:
        score += 2
        
    
    if sex =="Male":
        score=score+0
    elif sex == "Female":
        score=score+1
    
   
    if chf==False:
        score=score+0
    elif chf==True:
        score=score+1
        
    
    if hypertension==False:
        score=score+0
    elif hypertension==True:
        score=score+1
    
    if stroke==False:
        score=score+0
    elif stroke==True:
        score=score+2
        
        
    if vascular==False:
        score=score+0
    elif vascular==True:
        score=score+1
        
    
    if diabetes==False:
        score=score+0
    elif diabetes==True:
        score=score+1
    
    
    
    return score
    
    
````

## Part 2: HEART Score

The HEART score is a predictor for major cardiac events.  It requires 5 high-level inputs:
* History (Slightly / Moderately / Highly suspicious)
* EKG (Normal / Non-specific repolarization disturbance / Significant ST deviation)
* Age (Number)
* Risk Factors (Number of risk factors)
* Initial Troponin (Number of times the normal limit)

Fill out the function below with logic to calculate the numeric risk score for teh given input.

Be sure to provide meaningful documentation and at least two test cases in your documentation. Also make sure your code satisfies the test cases provided in the assert statements.

````python
def heart(history, ekg, age, risks, troponin):
    """(str,str,int,int,float)->int
    This function uses the logic from https://www.mdcalc.com/heart-score-major-cardiac-events
    to calculate the Heart score.
    
    >>>> ('Slightly suspicious','Non-specific repolarization',22,2,2.3)
    3
    
    >>>>('Slightly suspicious','Non-specific repolarization',32,4,1.3)
    4
    """
    ### YOUR SOLUTION HERE
    score=0
    
    if history =='Slightly suspicious':
        score=score+0
    elif history=='Moderately suspicious':
        score=score+1
    elif history== 'Highly suspicious':
        score=score+2

        
    if ekg == 'Normal':
        score=score+0
    elif ekg =='Non-specific repolarization':
        score=score+1
    elif ekg == 'Significant ST deviation':
        score=score+2

    
    if age <45:
        score=score+0
    elif (age>=45) and (age<=64):
        score=score+1
    elif age>= 65:
        score=score+2

    
    if risks == 0:
        score=score+0
    elif (risks>=1) and (risks<=2):
        score=score+1
    elif (risks>=3):
        score=score+2

    
    if troponin==0:
        score=score+0
    elif (troponin>=1) and (troponin<=3):
        score=score+1
    elif (troponin>3):
        score=score+2



    return score
    
        
````

## Part 3: Framingham Risk Score for Hard Coronary Heart Disease

The Framingham Risk Score for Hard Coronary Heart Disease is intended for non diabetic patients age 30-79 only.  So, if the patient's age is < 30 or > 79, your function should return `-1` rather than a specific risk score.

The Framingham Risk Score takes 7 inputs:
* Age (Number)
* Sex (Male / Female)
* Smoker (True / False)
* Total cholesterol (Number)
* HDL cholesterol (Number)
* Systolic BP (Number)
* Blood pressure being treated with medicines (No / Yes)

You'll not that rather than being a basic parametric equation, this is a regression function defined by coefficients.  It also requires you take the natural logarithm (`ln`) of many of the parameters.  To help you out, here's an example of how to interpret the formulat provided in the website's **Evidence** tab.

Take special note of the footnotes in the logic:
* *Yes=1, No=0 (for Treated for blood pressure and Smoker)
* ** Men: if age >70, use ln(70) x Smoker. Women: if age >78, use ln(78) x Smoker.


---

These segments of the equation and the coefficient table...

> $ L_{Men} = \beta \times \ln(Age) + \beta \times \ln(cholesterol) ... $
>
> | Variable &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; | Men | Women
> | :---------------------------- | :--: | :--: |
> | $ \ln(Age) $ | 52.00961 | 31.764001
> | $ \ln(Cholesterol) $ | 20.014077 | 22.465206

Could be written in Python as follows:

```python
import math

L = 0
if sex == "Male":
    L += 52.00961 * math.ln(age)
    L += 20.014077 * math.ln(cholesterol)

```

**ROUND YOUR FINAL RESULT TO 4 DECIMAL PLACES**

````python
import math
def framingham(age, sex, smoker, cholesterol, hdl, systolic, bp_treated):
    ### YOUR SOLUTION HERE
    """(int,str,bool,int,int,int,bool)->float
    This function uses the logic from https://www.mdcalc.com/framingham-risk-score-hard-coronary-heart-disease#evidence
    to calculate the score Framingham Risk Score for Hard Coronary Heart Disease.
    
    >>>>assert framingham(82,'Male',False,214,64,92,True) 
    -1
    
    >>>>assert framingham(36,'Male',True,169,33,128,True) 
    0.0465
    
    
    """
    #age
    score=0
    
    if age< 30 or age> 79:
        return -1
    
    L = 0
    if sex == "Male":
        L += 52.00961 * math.log(age)
        L += 20.014077 * math.log(cholesterol)
        L += -0.905964 * math.log(hdl)
        L += 1.305784 * math.log(systolic)
        L += -4.605038* math.log(cholesterol)*math.log(age)
        L += -2.93323 * math.log(age)*math.log(age)
        if  bp_treated:   
            L += 0.241549
        if smoker:
            L=L+12.096316 *1
            if age >70:
                L += -2.84367*math.log(70)

        L=L-172.300168
        return round(1 - 0.9402**math.exp(L),4) 
    else:
        L += 31.764001 * math.log(age)
        L += 22.465206 * math.log(cholesterol)
        L += -1.187731 * math.log(hdl)
        L += 2.552905 * math.log(systolic)
        L += -5.060998* math.log(cholesterol)*math.log(age)
        if bp_treated:
            L+=0.420251
        if smoker and age > 78:
            L += 13.07543* math.log(78)  
            L += -2.996945* math.log(smoker)*math.log(age)
        L=L-146.5933061
    
    
    
    
    return round(1-0.98767**math.exp(L),4)
    
````

````python
import csv
""" [str, float,float,float,bool]
Combining the part1, 2,3  into a list of lists.

>>>
""""





#cha2ds2_vasc(age, sex, chf, hypertension, stroke, vascular, diabetes): 
# """(int,str,bool,bool,bool,bool,bool)->int

# heart(history, ekg, age, risks, troponin):
#     """(str,str,int,int,float)->int

# framingham(age, sex, smoker, cholesterol, hdl, systolic, bp_treated):
#     """(int,str,bool,int,int,int,bool)->float
def normalize_gender(M_or_F):
    if M_or_F == 'M':
        return 'Male'
    else:
        return'Female'
    
def clean_yesno(bad_chf):
    if bad_chf == 'Yes':
        return True
    else:
        return False


    
    
def test_patients(filename):
    ### YOUR SOLUTION HERE
    with open(filename) as csvfile:    
        print(csvfile)
        patient=[]
        reader = csv.DictReader(csvfile)
        for row in reader:
            midterm_patient =(row['\ufeffpatient']) 
            

            cha_score=cha2ds2_vasc(int(row['age']), normalize_gender(row['sex']), clean_yesno(row["chf history"]), clean_yesno(row["hypertension history"]), clean_yesno(row["stroke history"]), clean_yesno(row["vascular disease history"]), clean_yesno(row["diabetes history"]))
           # print(cha_score)
            heart_score=heart((row['history']), row['ekg'], int(row['age']), int(row["risk factors"]), float(row["troponin"]))
            #print(heart_score)
            fram_score=framingham(int(row['age']), normalize_gender(row['sex']), clean_yesno(row["smoker"]), int(row['total cholesterol']), int(row['hdl cholesterol']), int(row['systolic bp']), clean_yesno(row['bp medicine']))
        
            total_score=(cha_score>=2 and heart_score>=4 and fram_score>=0.03)
            patient.append([midterm_patient,float(cha_score),float(heart_score),float(fram_score), total_score])
    print(patient)
    return patient
````
